Top 10 in Each Stat Per Season (PTS,TRB,AST,STL,BLK,TOV,3P%,FT%,FG%)

```python
import pandas as pd

df = pd.read_csv('NBA_Cleaned.csv') #load Cleaned Data

stats = ['PTS', 'TRB', 'AST', 'STL', 'BLK', 'TOV', '3P%', 'FT%', 'FG%']

# loop through each stat
for stat in stats:
    print(f"\n Top 10 Players in {stat} Per Season:")

    # loop through each season and create smaller df that only has rows from that season
    for season in df["Season"].unique():
        season_df = df[df['Season'] == season] 

        # Sort by stat descending and get top 10
        top10 = season_df.sort_values(by = stat,ascending = False).head(10)

        # Display the result
        print(f"\nðŸ”¹ Season {season} â€“ Top 10 in {stat}:")
        print(top10[['Player', 'Team', stat]].round(2))
```

Top 10 in Each Stat Overall (PTS,TRB,AST,STL,BLK,TOV,3P%,FT%,FG%)

```python
df = pd.read_csv('NBA_Cleaned.csv') #load Cleaned Data

stats = ['PTS', 'TRB', 'AST', 'STL', 'BLK', 'TOV', '3P%', 'FT%', 'FG%']


# Group the data by player name, and then calculate each player's average stats
player_stats = df.groupby('Player').agg({
    'PTS': 'mean',
    'TRB': 'mean',
    'AST': 'mean',
    'STL': 'mean',
    'BLK': 'mean',
    'TOV': 'mean',
    '3P%': 'mean',
    'FT%': 'mean',
    'FG%': 'mean',
    'G': 'sum'
})

# Filter: Only players with at least 300 games
player_stats = player_stats[player_stats['G'] >= 280]

# Drop 'G' column (not needed in final display)
player_stats = player_stats.drop(columns='G')

# Loop through each stat and print Top 10
for stat in stats:
    print(f"\n Top 10 Players by Average {stat} (Min 270 Games Played):")
    top10 = player_stats.sort_values(by = stat, ascending=False).head(10)
    print(top10[[stat]].round())
```

Top 10 in IMPACT Overall (PTS + TRB + AST + STL + BLK - TOV)


```python

df = pd.read_csv('NBA_Cleaned.csv')  # Load cleaned data

# Optional: Create an "Impact Score"
df['Impact'] = df['PTS'] + df['TRB'] + df['AST'] + df['STL'] + df['BLK'] - df['TOV']

# Stats you want to rank by
stats = ['PTS', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'Impact']

#Group by player: get average Impact and total games played

player_stats = df.groupby('Player').agg({
    'Impact': 'mean',
    'G': 'sum'
})

# Only keep players with at least 300 games played
player_stats = player_stats[player_stats['G'] >= 280]

# Drop games column (not needed in output)
player_stats = player_stats.drop(columns='G')

# Sort by impact and show top 10
top10_impact = player_stats.sort_values(by='Impact', ascending=False).head(10)

# Print result
print("\nðŸ”¥ Top 10 Players by Custom Impact Score (Min 270 Games Played):")
print(top10_impact.round(2))
```
